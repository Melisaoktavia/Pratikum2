```python
# ==============================
# ANALISIS CITRA DI GOOGLE COLAB
# ==============================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

def analyze_my_image():
    print("Silakan upload gambar utama...")
    uploaded = files.upload()

    image_path = list(uploaded.keys())[0]
    img = cv2.imread(image_path)

    if img is None:
        print("Gambar tidak ditemukan atau gagal dibaca!")
        return

    analysis_results = {}

    print("\n===== HASIL ANALISIS GAMBAR =====")

    # 1 Dimensi dan resolusi
    height, width, channels = img.shape
    analysis_results["width"] = width
    analysis_results["height"] = height
    analysis_results["channels"] = channels

    print("\n1. DIMENSI GAMBAR")
    print(f"Width  : {width} pixel")
    print(f"Height : {height} pixel")
    print(f"Channel: {channels}")

    # 2 Aspect Ratio
    aspect_ratio = width / height
    analysis_results["aspect_ratio"] = aspect_ratio
    print("\n2. ASPECT RATIO")
    print("Aspect Ratio :", aspect_ratio)

    # 3 Konversi ke grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("\n3. UKURAN SETELAH GRAYSCALE")
    print("Shape grayscale:", gray.shape)

    # 4 Statistik citra
    mean_val = np.mean(img)
    std_val = np.std(img)
    min_val = np.min(img)
    max_val = np.max(img)

    analysis_results["mean"] = mean_val
    analysis_results["std"] = std_val
    analysis_results["min"] = min_val
    analysis_results["max"] = max_val

    print("\n4. STATISTIK CITRA")
    print("Mean :", mean_val)
    print("Std  :", std_val)
    print("Min  :", min_val)
    print("Max  :", max_val)

    # 5 Histogram RGB
    print("\n5. HISTOGRAM RGB")
    colors = ('b', 'g', 'r')
    plt.figure()
    for i, col in enumerate(colors):
        hist = cv2.calcHist([img], [i], None, [256], [0,256])
        plt.plot(hist)
    plt.title("Histogram RGB")
    plt.xlabel("Intensitas Pixel")
    plt.ylabel("Jumlah Pixel")
    plt.show()

    # 6 Perbandingan dengan gambar sample (opsional)
    print("\nApakah ingin membandingkan dengan gambar sample? (y/n)")
    choice = input()

    if choice.lower() == 'y':
        print("Silakan upload gambar sample...")
        uploaded_sample = files.upload()
        sample_path = list(uploaded_sample.keys())[0]
        sample_img = cv2.imread(sample_path)

        if sample_img is not None:
            sample_mean = np.mean(sample_img)
            print("\n6. PERBANDINGAN")
            print("Mean gambar utama  :", mean_val)
            print("Mean gambar sample :", sample_mean)
        else:
            print("Sample gagal dibaca.")

    return analysis_results


# Jalankan fungsi
results = analyze_my_image()

print("\n===== DATA DICTIONARY =====")
print(results)
```
