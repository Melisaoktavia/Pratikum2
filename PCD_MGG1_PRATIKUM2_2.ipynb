```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_digitization(analog_signal, sampling_rate, quantization_levels):

    # 1. Buat domain kontinu
    x_cont = np.linspace(0, 1, 1000)
    y_cont = analog_signal(x_cont)

    # 2. Sampling
    x_sample = np.linspace(0, 1, sampling_rate)
    y_sample = analog_signal(x_sample)

    # 3. Quantization
    y_min = np.min(y_sample)
    y_max = np.max(y_sample)

    # Skala ke 0 - quantization_levels
    y_quant = np.round(
        (y_sample - y_min) / (y_max - y_min) * (quantization_levels - 1)
    )

    # Kembalikan ke skala asli
    y_quant = y_quant / (quantization_levels - 1) * (y_max - y_min) + y_min

    # 4. Visualisasi
    plt.figure()
    plt.plot(x_cont, y_cont, label="Analog Signal")
    plt.stem(x_sample, y_quant, linefmt='r-', markerfmt='ro', basefmt=" ")
    plt.title("Simulasi Digitalisasi")
    plt.legend()
    plt.show()

    return x_sample, y_quant


# Contoh penggunaan
analog_function = lambda x: np.sin(2 * np.pi * 5 * x)

simulate_digitization(
    analog_signal=analog_function,
    sampling_rate=20,
    quantization_levels=8
)
```
